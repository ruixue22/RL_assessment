### Digital Talents

```python
# get digital talents
digital_talent_df['new_date'] = pd.to_datetime(digital_talent_df['Season Premiere Date'])
digital_time_frame = digital_talent_df[(digital_talent_df['new_date'] > '10/1/21') 
                                       & (digital_talent_df['new_date'] < '7/1/22')] #the 9 month
digital_time_frame = digital_time_frame.sort_values(by='new_date')
digital_time_frame = digital_time_frame.rename(columns = {'Platform':"Distributor/Platform/Network"})

digital_time_frame
```

```python
# get tv talents only
tv_talent_df['new_date'] = pd.to_datetime(tv_talent_df['Season Premiere Date'])
tv_time_frame = tv_talent_df[(tv_talent_df['new_date'] > '10/1/21') & (tv_talent_df['new_date'] < '6/30/22')]
tv_time_frame = tv_time_frame.sort_values(by='new_date')
tv_time_frame = tv_time_frame.rename(columns = {'Network':"Distributor/Platform/Network"})

tv_time_frame
```

```python
# get film talents
film_talent_df['new_date'] = pd.to_datetime(film_talent_df['U.S. release date for films'])
film_time_frame = film_talent_df[(film_talent_df['new_date'] > '10/1/21') & (film_talent_df['new_date'] < '7/1/22')]
film_time_frame = film_time_frame.sort_values(by='new_date')
film_time_frame = film_time_frame.rename(columns = {'Distributor':"Distributor/Platform/Network"})
film_time_frame

```

### Add "Product Type" attribute

```python
film_time_frame = film_time_frame.assign(Project_Type = "FILM")

# shift column 'Name' to first position
first_column = film_time_frame.pop('Project_Type')
  
# insert column using insert(position,column_name,
# first_column) function
film_time_frame.insert(0, 'Project_Type', first_column)
film_time_frame
```

```python
digital_time_frame = digital_time_frame.assign(Project_Type = "EPISODIC(digital)")

# shift column 'Name' to first position
first_column = digital_time_frame.pop('Project_Type')
  
# insert column using insert(position,column_name,
# first_column) function
digital_time_frame.insert(0, 'Project_Type', first_column)
digital_time_frame
```

```python
tv_time_frame = tv_time_frame.assign(Project_Type = "EPISODIC(tv)")

# shift column 'Name' to first position
first_column = tv_time_frame.pop('Project_Type')
  
# insert column using insert(position,column_name,
# first_column) function
tv_time_frame.insert(0, 'Project_Type', first_column)
tv_time_frame
```

### Finalize Data Cleaning:
1. Combine tv+digital for EPISODIC dataframe
2. Filter Series Regular and Cast for both EPISODIC and FILM 

```python
# combine the TV data frames:
tv_frames = [tv_time_frame, digital_time_frame]  
TV_df = pd.concat(tv_frames)
TV_df = TV_df[TV_df['Job Title'] == 'Series Regular']
display(TV_df)

# print(tv_time_frame.shape)
# print(digital_time_frame.shape)
# print(TV_df.shape)

# set the final name for film dataframe:
FILM_df = film_time_frame
FILM_df = FILM_df[FILM_df['Job Title'] == 'Cast']
display(FILM_df)
```

```python
# remove unnecessary columns

FILM_df = FILM_df[['Project_Type', 'VI Project ID', 'Title',
       'Distributor/Platform/Network', 'U.S. release date for films',
       'Season Premiere Date', 'Genre', 'Arena', 'Episodic Order', 'Job Title', 'VI talent ID',
       'Name', 'Gender New', 'Non Binary Details', 'Ethnicity New',
       '2+ Ethnicity', 'new_date']]
print(FILM_df.columns)

TV_df = TV_df[['Project_Type', 'VI Project ID', 'Title', 'Distributor/Platform/Network', 'U.S. release date for films',
       'Season Premiere Date', 'Genre', 'Arena', 'Episodic Order', 'Season Number', 'Job Title',
       'VI talent ID', 'Name', 'Gender New', 'Non Binary Details',
       'Ethnicity New', '2+ Ethnicity','new_date']]
print(TV_df.columns)
```

### Penetration Rate Calculations: FILMS

```python
# display(FILM_df[FILM_df['Ethnicity New'] == 'Black'])
# display(FILM_df[FILM_df['2+ Ethnicity'].str.contains("Black", na = False)])

film_black = FILM_df[FILM_df['Ethnicity New'] == 'Black']
film_2black = FILM_df[FILM_df['2+ Ethnicity'].str.contains("Black", na = False)]

# print(film_black.shape, film_2black.shape)

film_all_black = pd.concat([film_black,film_2black])
display(film_all_black)
```

```python
# Checking that only black talents were selected
film_all_black['Job Title'].unique()
```

```python
# FILM RESULTS
black_cast_count = film_all_black.shape[0]
all_cast_count = FILM_df.shape[0]

black_film_penetration = black_cast_count / all_cast_count

FILM_pecentage = pd.DataFrame([[black_cast_count, all_cast_count, black_film_penetration]], 
             columns = ['Black Cast Count' , 'All Cast Count', 'Black Cast Penetration'])
FILM_pecentage.T
```
